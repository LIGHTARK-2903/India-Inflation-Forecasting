```python
# EDA STEP 1: imports + load cleaned CPI & WPI (run this first)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

pd.set_option('display.precision', 2)

# Confirm working directory and files
import os
print("Working dir:", os.getcwd())
print("Files in data/:", os.listdir("data"))

# Load cleaned datasets
cpi = pd.read_csv("data/cpi_clean.csv", index_col=0, parse_dates=True).squeeze("columns")
wpi = pd.read_csv("data/wpi_clean.csv", index_col=0, parse_dates=True).squeeze("columns")

print("\nLoaded series:")
print("CPI:", cpi.index.min(), "→", cpi.index.max(), "| rows:", len(cpi))
print("WPI:", wpi.index.min(), "→", wpi.index.max(), "| rows:", len(wpi))
```

```python
# EDA STEP 2: CPI vs WPI Trend Plot

plt.figure(figsize=(12,4))
plt.plot(cpi.index, cpi.values, label='CPI (Index)', linewidth=2)
plt.plot(wpi.index, wpi.values, label='WPI (Index)', linewidth=1.5, alpha=0.8)

plt.title("CPI vs WPI — Monthly Index (2012–2025)", fontsize=14)
plt.ylabel("Index Level")
plt.grid(alpha=0.25)
plt.legend()

plt.tight_layout()
plt.savefig("plots/trend_cpi_wpi.png", dpi=150)
plt.show()
```

```python
# EDA STEP 3: Compute MoM% and YoY% inflation

# Month-on-Month: (current - previous) / previous * 100
cpi_mom = cpi.pct_change() * 100
wpi_mom = wpi.pct_change() * 100

# Year-on-Year: (current - last year same month) / last year * 100
cpi_yoy = cpi.pct_change(12) * 100
wpi_yoy = wpi.pct_change(12) * 100

# Display the last 6 values for quick check
display(pd.DataFrame({
    "CPI_MoM (%)": cpi_mom.tail(6),
    "CPI_YoY (%)": cpi_yoy.tail(6),
    "WPI_YoY (%)": wpi_yoy.tail(6)
}))
```

```python
# EDA STEP 4: YoY Inflation Plot

plt.figure(figsize=(12,4))
plt.plot(cpi_yoy.index, cpi_yoy, label="CPI YoY (%)", linewidth=2)
plt.plot(wpi_yoy.index, wpi_yoy, label="WPI YoY (%)", linewidth=1.5, alpha=0.8)

plt.axhline(0, color='black', linewidth=0.7)
plt.title("Year-on-Year Inflation: CPI vs WPI", fontsize=14)
plt.ylabel("YoY Inflation (%)")
plt.legend()
plt.grid(alpha=0.25)

plt.tight_layout()
plt.savefig("plots/yoy_cpi_wpi.png", dpi=150)
plt.show()
```

```python
# EDA STEP 5: Decompose CPI (Trend / Seasonal / Residual)

# Remove NA values
cpi_cleaned = cpi.dropna()

# Multiplicative is appropriate if the series grows over time
model = 'multiplicative' if (cpi_cleaned > 0).all() else 'additive'

result = seasonal_decompose(
    cpi_cleaned,
    model=model,
    period=12,
    extrapolate_trend='freq'
)

fig = result.plot()
fig.set_size_inches(12, 8)
fig.suptitle("CPI Decomposition: Trend, Seasonality, Residuals", fontsize=16)

plt.tight_layout()
plt.savefig("plots/cpi_decomposition.png", dpi=150)
plt.show()
```

```python
# EDA STEP 6: Rolling averages for CPI YoY

# 3-month moving average (short-term momentum)
cpi_yoy_ma3 = cpi_yoy.rolling(3).mean()

# 12-month moving average (long-term inflation trend)
cpi_yoy_ma12 = cpi_yoy.rolling(12).mean()

plt.figure(figsize=(12,4))
plt.plot(cpi_yoy.index, cpi_yoy, label="CPI YoY (raw)", alpha=0.35)
plt.plot(cpi_yoy_ma3.index, cpi_yoy_ma3, label="3M Moving Avg", linewidth=1.8)
plt.plot(cpi_yoy_ma12.index, cpi_yoy_ma12, label="12M Moving Avg", linewidth=2)

plt.title("CPI YoY Inflation — Short & Long-Term Trends", fontsize=14)
plt.ylabel("Inflation (%)")
plt.legend()
plt.grid(alpha=0.25)

plt.tight_layout()
plt.savefig("plots/cpi_yoy_moving_avg.png", dpi=150)
plt.show()
```

```python
# EDA STEP 7: CPI vs WPI Correlation (YoY inflation)

# Build correlation dataframe
df_corr = pd.concat([cpi_yoy, wpi_yoy], axis=1)
df_corr.columns = ["CPI_YoY", "WPI_YoY"]

# Compute correlation
corr_val = df_corr.corr().iloc[0,1]
print("Correlation between CPI YoY and WPI YoY =", round(corr_val, 3))

# Heatmap
plt.figure(figsize=(4,4))
sns.heatmap(df_corr.corr(), annot=True, fmt=".2f", vmin=-1, vmax=1, cmap="coolwarm")
plt.title("Correlation Heatmap (CPI YoY vs WPI YoY)")
plt.tight_layout()
plt.savefig("plots/cpi_wpi_correlation.png", dpi=150)
plt.show()
```

```python
# EDA STEP 8: Cross-correlation to test spatial lead-lag relationship

def cross_corr(a, b, max_lag=12):
    """
    Computes correlation of a vs b at different lags.
    Positive lag: CPI leads WPI
    Negative lag: WPI leads CPI
    """
    results = {}
    for lag in range(-max_lag, max_lag + 1):
        if lag < 0:
            # WPI leads by -lag months
            results[lag] = a.corr(b.shift(-lag))
        else:
            # CPI leads by lag months
            results[lag] = a.corr(b.shift(lag))
    return pd.Series(results)

cc = cross_corr(cpi_yoy.dropna(), wpi_yoy.dropna(), max_lag=12)

plt.figure(figsize=(10,4))
plt.stem(cc.index, cc.values, basefmt=" ")
plt.axhline(0, color='black', linewidth=0.6)
plt.title("Cross-Correlation: CPI YoY vs WPI YoY\n(Negative lag → WPI leads CPI)", fontsize=12)
plt.xlabel("Lag (months)")
plt.ylabel("Correlation")
plt.tight_layout()
plt.savefig("plots/crosscorr_cpi_wpi.png", dpi=150)
plt.show()

# Identify strongest lead/lag
best_lag = cc.idxmax()
best_corr = cc.max()

print("Max correlation at lag =", best_lag, "| correlation =", round(best_corr, 3))
```

```python
# EDA STEP 9: ACF & PACF for CPI YoY

plt.figure(figsize=(10,3))
plot_acf(cpi_yoy.dropna(), lags=36, zero=False)
plt.title("ACF — CPI YoY")
plt.tight_layout()
plt.savefig("plots/acf_cpi_yoy.png", dpi=150)
plt.show()

plt.figure(figsize=(10,3))
plot_pacf(cpi_yoy.dropna(), lags=36, zero=False, method='ywm')
plt.title("PACF — CPI YoY")
plt.tight_layout()
plt.savefig("plots/pacf_cpi_yoy.png", dpi=150)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA(1,1,1) to CPI
model_arima = ARIMA(cpi, order=(1, 1, 1))
result_arima = model_arima.fit()

print(result_arima.summary())
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit SARIMA(1,1,1)(1,1,1,12)
model_sarima = SARIMAX(
    cpi,
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

result_sarima = model_sarima.fit()

print(result_sarima.summary())
```

```python
# Stationarity tests
import statsmodels.tsa.stattools as ts

# Augmented Dickey-Fuller
adf_res = ts.adfuller(cpi.dropna())
print("ADF p-value:", adf_res[1])

# KPSS (null = stationary)
kpss_res = ts.kpss(cpi.dropna(), nlags="auto")
print("KPSS p-value:", kpss_res[1])

# Interpretation hint:
# - ADF p-value < 0.05 => reject unit root => stationary
# - KPSS p-value < 0.05 => reject stationarity (i.e., non-stationary)
```

```python
# Quick diagnostic plots & Ljung-Box test for the fitted SARIMA
result_sarima.plot_diagnostics(figsize=(12,10))
plt.tight_layout()
plt.show()

# Ljung-Box on residuals (up to 24 lags)
from statsmodels.stats.diagnostic import acorr_ljungbox
lb = acorr_ljungbox(result_sarima.resid.dropna(), lags=[6,12,18,24], return_df=True)
print("Ljung-Box p-values:\n", lb["lb_pvalue"])
```

```python
import pmdarima as pm

auto = pm.auto_arima(
    cpi,
    start_p=0, max_p=3,
    start_q=0, max_q=3,
    d=None,           # let the algorithm decide
    seasonal=True,
    start_P=0, max_P=2,
    start_Q=0, max_Q=2,
    D=None,           # let it determine seasonal differencing
    m=12,
    stepwise=True,
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    information_criterion='aic'
)

print("Auto ARIMA summary:")
print(auto.summary())
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.statespace.sarimax import SARIMAX
import numpy as np

def metrics(y_true, y_pred):
    mse  = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae  = mean_absolute_error(y_true, y_pred)
    mape = (abs((y_true - y_pred)/y_true)).mean()*100
    return {"RMSE": rmse, "MAE": mae, "MAPE": mape}

h = 12
train = cpi.iloc[:-h]
test = cpi.iloc[-h:]

# --- Manual SARIMA ---
manual_mod = SARIMAX(
    train,
    order=(1,1,1),
    seasonal_order=(1,1,1,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)
manual_res = manual_mod.fit(disp=False)
manual_pred = manual_res.get_forecast(steps=h).predicted_mean

# --- Auto SARIMA ---
auto_mod = SARIMAX(
    train,
    order=(0,1,1),
    seasonal_order=(1,0,1,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)
auto_res = auto_mod.fit(disp=False)
auto_pred = auto_res.get_forecast(steps=h).predicted_mean

print("Manual SARIMA metrics:", metrics(test, manual_pred))
print("Auto SARIMA metrics:", metrics(test, auto_pred))
```

```python
# Fit final model on full CPI data
final_model = SARIMAX(
    cpi,
    order=(0,1,1),
    seasonal_order=(1,0,1,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

final_res = final_model.fit(disp=False)
print(final_res.summary())
```

```python
# FORECAST STEP 1: Forecast next 12 months with CI, save and plot
h = 12  # forecast horizon in months

# Forecast
fc = final_res.get_forecast(steps=h)
pred_mean = fc.predicted_mean
pred_ci = fc.conf_int(alpha=0.05)  # 95% CI

# Build a DataFrame for output
forecast_df = pd.DataFrame({
    "CPI_forecast": pred_mean,
    "CPI_lower_95": pred_ci.iloc[:, 0],
    "CPI_upper_95": pred_ci.iloc[:, 1]
})
forecast_df.index.name = "Date"

# Save
forecast_df.to_csv("outputs/cpi_forecast_12m.csv")

# Plot: historical + forecast + CI
plt.figure(figsize=(12,5))
plt.plot(cpi.index, cpi, label="CPI (observed)", linewidth=1.6)
plt.plot(forecast_df.index, forecast_df["CPI_forecast"], label=f"CPI forecast ({h}m)", linewidth=2, color="C1")
plt.fill_between(forecast_df.index,
                 forecast_df["CPI_lower_95"],
                 forecast_df["CPI_upper_95"],
                 color="C1", alpha=0.18, label="95% CI")
plt.axvline(cpi.index.max(), color='k', linestyle='--', alpha=0.6)
plt.title(f"CPI: Historical and {h}-month Forecast (SARIMA final model)")
plt.legend()
plt.grid(alpha=0.2)
plt.tight_layout()
plt.savefig("plots/cpi_forecast_12m.png", dpi=200)
plt.show()

print("Saved outputs/cpi_forecast_12m.csv — rows:", len(forecast_df))
display(forecast_df)
```

```python
# FORECAST STEP 2: Compute YoY % inflation for forecast months
combined = pd.concat([cpi, forecast_df["CPI_forecast"]]).rename("CPI")
combined = combined.sort_index()

combined_yoy = combined.pct_change(12) * 100
forecast_yoy = combined_yoy.loc[forecast_df.index]

forecast_yoy_df = pd.DataFrame({
    "CPI_forecast_index": forecast_df["CPI_forecast"],
    "CPI_forecast_YoY_pct": forecast_yoy.values,
    "CPI_lower_95": forecast_df["CPI_lower_95"],
    "CPI_upper_95": forecast_df["CPI_upper_95"]
}, index=forecast_df.index)

forecast_yoy_df.to_csv("outputs/cpi_forecast_12m_yoy.csv")

display(forecast_yoy_df.head(3))
print("Saved outputs/cpi_forecast_12m_yoy.csv")
```

```python
# FORECAST STEP 4: Summary of next-12m YoY range
forecast_yoy_df = pd.read_csv("outputs/cpi_forecast_12m_yoy.csv", index_col=0, parse_dates=True)
min_yoy = forecast_yoy_df["CPI_forecast_YoY_pct"].min()
max_yoy = forecast_yoy_df["CPI_forecast_YoY_pct"].max()
mean_yoy = forecast_yoy_df["CPI_forecast_YoY_pct"].mean()

print(f"Forecast YoY % range over next {len(forecast_yoy_df)} months: min={min_yoy:.2f}%, mean={mean_yoy:.2f}%, max={max_yoy:.2f}%")
```
